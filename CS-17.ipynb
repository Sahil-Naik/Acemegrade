### 1. You are given a dataset, containing the number of hurricanes (Hurricanes.csv) occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

hurr_df = pd.read_csv("datasets/Hurricanes.csv")

plt.figure(figsize = (15,10))
plt.bar(hurr_df['Year'],hurr_df['Hurricanes'])
plt.show()
```

### 4.3 configure the axes.

```python
import matplotlib.pyplot as plt

# given data points
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

plt.figure(figsize = (10,8))
plt.plot(x, y, linestyle='--', marker='o', color='green')

plt.xlim(-3,6)
plt.xticks([-2,-1,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])

plt.ylim(19.0,22.0)
plt.yticks([19.5,20,20.5,21,21.5,22,22.5,23])
plt.show()
```

### 4.4 Give title of Graph & labels of x axis and y axis.

```python
import matplotlib.pyplot as plt

# given data points
x = [1,2,3,4]
y = [200, 205, 216, 212]

plt.figure(figsize = (10,8))
plt.plot(x, y, linestyle='--', marker='*', color='black')

plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])

plt.ylim(190,230)
plt.yticks([195,200,205,210,215,220])

plt.xlabel('X data')
plt.ylabel('Y data')
plt.title("Simple Plot")

plt.show()
```

### 4.5 Give error bar if y_error = [0.12, 0.13, 0.2, 0.1].

```python
import matplotlib.pyplot as plt

# given data points
x = [1,2,3,4]
y = [40,49,42,53.6]

# error data
y_error = [0.9,1.2,0.5,1]

plt.figure(figsize = (10,8))
plt.plot(x, y, linestyle='-.', marker='o', color='green')
plt.errorbar(x, y, yerr=y_error, linestyle="none", marker="None", color="red")

plt.title('Simple Plot')
plt.xlabel("X data")
plt.ylabel("Y data")

plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])

plt.ylim(39,55)
plt.yticks([40,49,42,53.6])


plt.show()
```

### 4.8 Draw a scatter graph of any 50 random values of x and y axis.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50

x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

plt.figure(figsize = (20,20))
area = np.pi * (15 * np.random.rand(N))**3
plt.scatter(x, y, s=area*2, c=colors, alpha=0.5)

plt.show()
```

### 4.9 Create a dataframe from following data:
**'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy']<br>
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze']<br>
'female': [0, 1, 1, 0, 1]<br>
'age': [42, 52, 36, 24, 73]<br>
'preTestScore': [4, 24, 31, 2, 3]<br>
'postTestScore': [25, 94, 57, 62, 70]<br>
Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age.<br>**

```python
import matplotlib.pyplot as plt

data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
        'female': [0, 1, 1, 0, 1],
        'age': [42, 52, 36, 24, 73],
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])

print("Given data is:\n{}\n\n".format(df.head()))

plt.figure(figsize=(4,5))
plt.scatter(df.preTestScore, df.postTestScore, s=df.age)
plt.show()
```

### 4.10 Draw a Scatterplot from the data in question 4.9 of preTestScore and postTestScore with the size = 300 and the color determined by sex.

```python
plt.figure(figsize=(4,5))
plt.scatter(df.preTestScore, df.postTestScore, s=300, c=df.female)
plt.show()
```
