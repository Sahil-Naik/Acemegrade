## Data Manipulation

### Create a new column - 'total_benefitted' that is a sum of inmates benefitted through all modes.

```python
pris['total_benefitted'] = pris.iloc[:,2:].sum(axis=1)
pris
```

---

### Create a new row - 'totals' that is the sum of all inmates benefitted through each mode across all states.

```python
pris_total = pris.append(pris.iloc[:,2:].sum(numeric_only=True), ignore_index=True)
pris_total
```

---

## Plotting

### Make a bar plot with each state name on the x-axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
plot.figure(figsize = (15,8))
xlabels = pris['STATE/UT'].values

plot.xticks(np.arange(xlabels.shape[0]), xlabels, rotation='vertical', fontsize=12)
plot.bar(np.arange(pris.values.shape[0]), pris['total_benefitted'], align='edge')
plot.show()
```

---

### Make a pie chart that depicts the ratio among different modes of benefits.

```python
fig1, ax1 = plot.subplots()

labels = pris.columns[2:-1]
sizes = pris_total.iloc[-1][2:-1].values

ax1.pie(sizes, labels=labels, autopct="%1.1f%%")
ax1.axis('equal')
plot.show()
```
