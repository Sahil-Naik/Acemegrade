```python
x = df[['Avg_Session_Length', 'Time_on_App',
        'Time_on_Website', 'Length_of_Membership']]
x
```

```python
y = df['Yearly_Amount_Spent']
y
```

### Train-Test splits

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=85)
```

### Creating model

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(X_train, y_train)
```

### Prediction and result as graph

```python
predicts = LR.predict(X_test)
plt.scatter(y_test, predicts)
plt.xlabel('Y-Test')
plt.ylabel('Predicted-Y')
```

### Coefficient and Intercept

```python
LR.coef_
```

```python
LR.intercept_
```

### Root mean squared error

```python
from sklearn import metrics

rmse = np.sqrt(metrics.mean_squared_error(y_test, predicts))
print("RMSE: {}".format(round(rmse,3)))
```

### DIST plot

```python
sns.displot((y_test, predicts), bins=50)
```

###  Q. Based on coefficients interpret company should focus more on their mobile app or on their website.

```python
coef = pd.DataFrame(LR.coef_, x.columns)
coef.columns = ['Coeffecient']
coef
```

### We can interpret the above data as:

1. The Avg_Session_Length is associated with an increase of 26.081 total dollars spent per 1 unit.

2. The Time_on_app is associated with an increase of 39.187 total dollars spent per 1 unit.

3. The Time_on_website is associated with an increase of 0.404 total dollars spent per 1 unit.

4. The Length_of_Membership is associated with an increase of 61.417 total dollars spent per 1 unit.

### Hence the company should focus more on their mobile app and not on their website.
