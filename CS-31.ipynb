### Q2. Replace the missing values by the most frequent value in each column. 

```python
Null_Val = pd.concat([df.isnull().sum(),df.dtypes], axis = 1)
Null_Val.columns= ['Null_Count', 'Datatype']
Null_Val
```

### Columns with missing values and data types

```python
category_cols= Null_Val[(Null_Val.Null_Count>0)&(Null_Val.Datatype=='object')]
category_cols
```

```python
category_variables = list(category_cols.index)
category_variables
```

### Update missing values

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="most_frequent")
df[category_variables]= imp.fit_transform(df[category_variables])
df
```

### Checking missing value count

```python
Null_Val = pd.concat([df.isnull().sum(),df.dtypes], axis = 1)
Null_Val.columns= ['Null_Count', 'Datatype']
Null_Val
```

### Save result

```python
df.to_excel("datasets/processed/CS32.xlsx")
```

### Float columns with missing record count

```python
float_cols= Null_Val[(Null_Val.Null_Count>0)]
float_cols
```

```python
float_variables = list(float_cols.index)
float_variables
```

### Update missing values for float column

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="mean")
df[float_variables]= imp.fit_transform(df[float_variables])
```

```python
df.to_excel("datasets/processed/CS32.xlsx")
```

```python
df
```

### Q3. Data  contains many categorical features, replace them with label encoding.

```python
# Getting all columns list with data-type Object

All_cats = Null_Val[Null_Val.Datatype=='object']
All_cats = list(All_cats.index)
All_cats
```

### Hot Encoding

```python
new_df = df
for i in All_cats:
    new_df=pd.get_dummies(new_df, columns=[i], drop_first=True)
    
new_df.to_excel("datasets/processed/CS32_one_hot.xlsx")
new_df
```

```python
OC.unique()
```

### Label encoding for target column

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
target = label.fit_transform(OC)
print(target)
```

---

### Q4. Fit a decision tree classifier and observe the accuracy.

```python
# Train-Test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_df, target, test_size=0.2, random_state=1)

print("X_train shape = {}.".format(X_train.shape))
print("X_test shape = {}.".format(X_test.shape))
```

### Decision tree model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

### Predictions on X_test

```python
ydt_pred = dt.predict(X_test)
ydt_pred
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ydt_pred, y_test)
print(accuracy)
```

---

### Q5. Fit a random forest classifier and observe the accuracy.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

### Predictions on X_test

```python
yrf_pred = rf.predict(X_test)
yrf_pred
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(yrf_pred, y_test)
print(acc)
```
