### Importing libraries and loading data

```python
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = Image.open("images/dogs.jpeg")
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img)
```

### Creating array for image

```python
image_np = np.asarray(img)
```

### Reshpae the array

```python
image_np.shape
image_np = image_np.reshape((185*272,3))
image_np.shape
```

### Elbow method

```python
from sklearn.cluster import KMeans 
wcss = []
for i in range(1,11):
    kmeans = KMeans (n_clusters =  i , init = 'k-means++')
    kmeans.fit(image_np)
    wcss.append (kmeans.inertia_)
    
wcss
```

### Plot the graph

```python
plt.plot (range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel ('wcss')
```

### Creating model

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(image_np)
```

### Displaying the image

```python
labels = kmeans.predict(image_np)
labels = labels.reshape((185,272))
plt.imshow(labels)
```

### Displaying first colour

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
blob = np.ones((100,100,3))
blob1 = blob * centers[0]
plt.imshow(blob1.astype(np.uint8))
```

### Displaying second colour

```python
blob2 = blob * centers[1]
plt.imshow(blob2.astype(np.uint8))
```

### Displaying colour code as array

```python
blob2=blob2.astype(int)
blob2
```

### Displaying third colour

```python
blob3 = blob * centers[2]
plt.imshow(blob3.astype(np.uint8))
```
